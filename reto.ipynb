```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import warnings
import numpy as np
```

# Leemos el CSV y realizamos preprocesamiento

Comenzamos el desarrollo mediante un preprocesamiento de datos, en el que deberemos leer el fichero CSV correspondiente que ese encuentra en la carpeta data y tiene el nombre de all_2022.csv.
Posteriormente, debemos eliminar todos aquellos datos con valor nulo, pues resultan ruido en el estudio, todo ello lo guardamos en el dataset llamado datasetNA

```python
warnings.filterwarnings("ignore")
datasetOG = pd.read_csv("data/all_2022.csv")
datasetNA = datasetOG.dropna()
datasetNA.head(30)
```

# Limpiamos el dataset 

Una vez leído y eliminados los datos nulos del dataset, elaboraremos un nuevo dataset con los datos que nos resulten útiles, de esta manera estaremos operando únicamente con los datos necesarios y mejoraremos el rendimiento del análisis ocupando menos tiempo a la hora de realizar operaciones.

Para ello decidimos mantener las columnas más relevantes del dataset original llamado datasetNA y lo llevamos a un nuevo dataset llamado datasetAG que contara con las columnas de 'date', 'contract', 'buy_agent' , 'sell_agent', 'price', 'energy' ,'contract_start' y 'contract_end', posteriormente lo agruparemos por agente que ha realizado la operación de compra.

```python
#Eliminamo datos nulos conflictivos
#Nos quedamos con las columnas que nos interesen
datasetAG = datasetNA[['date', 'contract', 'buy_agent' , 'sell_agent', 'price', 'energy' ,'contract_start' , 'contract_end']]
datasetAG.groupby('buy_agent')
datasetAG.head(50)
```

# Calculamos el Vmap de cada empresa
----

Una vez obtenido el dataset objetivo deberemos calcular el Vwap perteneciente a cada agente que ha realizado una operación, para ello debemos obtener la suma total del volumen de ventas realizado por el agente, del mismo modo que el sumatorio de precio por el volumen de la venta realizada, de esta manera podemos obtener el Volume-Weighted Average Price lo que nos permitirá saber el precio promedio de compra de un activo por cada agente.

Para obtener estos datos y poder organizarlos de manera estructurada elaboramos un nuevo dataset con nombre DatasetAnalizar que contara con 3 columnas:

- La columna volume que contara con la suma del volumen total de todas las operaciones realizadas
- La columna total_price_energy que contara con el precio total de las operaciones realizadas por su volumen
- La columna vwap donde se calcualará el Volume-Weighted Average Price

```python
#DatasetAnalizar = datasetAG['price'] * datasetAG['energy']
# Calcular el sumatorio de energy por buy_agent
DatasetAnalizar = datasetAG.groupby('buy_agent')['energy'].sum().reset_index()
DatasetAnalizar['total_price_energy'] = datasetAG['price'] * datasetAG['energy']
DatasetAnalizar = DatasetAnalizar.dropna()
DatasetAnalizar = DatasetAnalizar.rename(columns={'energy': 'volume'})

DatasetAnalizar['vwap'] = DatasetAnalizar['total_price_energy'] / DatasetAnalizar['volume'] 
DatasetAnalizar = DatasetAnalizar.sort_values('vwap', ascending=False)
DatasetAnalizar.head(50)
# Imprimir el nuevo dataset
```

# Añadimos una puntaucion a cada agent
----
Una vez obtenido el vmap de cada agente debemos asignarle una puntuación en función de cada operación realizada y en base al valor vmap obtenido, tal y como se menciona alguien que pudiera vender estructuralmente por encima de vwap, o comprar por debajo, estaría haciendo un buen trabajo, por lo tanto, le otorgaremos un punto, realizado a través de un contador.

Creamos una nueva columna en el dataset llena de 0 que iremos cambiando en función del valor.

```python
DatasetAnalizar['puntuacion'] = 0
contador = 0
```

```python
buy_agents = ['RENEW', 'ZERE', 'ISMSL', 'CEPEN', 'ELVA2', 'PORTU', 'JAFP', 'SPBB', 'ENARA', 'PESL', 'MOLE', 'OESTE', 'YELLO', 'NOBE', 'VIEE', 'CONEV', 'VINI', 'GOIEN', 'WWHOL', 'LUZBO', 'GRTEN', 'GSLZE', 'LOBO', 'NEOEN', 'EEXXI', 'UMEME', 'POTEN', 'GNSUR', 'PEUSA', 'LUSIA', 'HAPO', 'NCE', 'TEVI', 'INSIG', 'POWER', 'CZFV', 'BSGEA', 'MODI', 'LAUEI', 'VISAL', 'LOGIC', 'COGEN', 'VIVA', 'LUZSE', 'LUVON', 'MULTI', 'CTNU', 'RECON', 'RENTR', 'GVME']
vwaps = [66.316667, 21.468293, 15.110588, 14.375610, 12.132000, 6.962312, 2.122615, 1.844895, 1.728205, 1.677311, 1.634775, 1.625097, 1.621094, 1.576404, 1.409660, 1.368337, 1.285853, 1.210068, 1.207400, 1.117952, 0.718724, 0.715610, 0.695116, 0.693617, 0.618039, 0.492107, 0.465765, 0.378509, 0.361134, 0.326656, 0.307249, 0.270914, 0.263590, 0.256495, 0.248362, 0.213495, 0.205934, 0.203934, 0.179798, 0.178831, 0.173025, 0.169663, 0.157018, 0.133511, 0.121348, 0.119382, 0.117403, 0.089885, 0.086173, 0.080551]
i = 0
for buy_agent in buy_agents:
    dataset_buy_agent_b = datasetAG[datasetAG['buy_agent'] == buy_agent]
    dataset_buy_agent_s = datasetAG[datasetAG['sell_agent'] == buy_agent]
    dataset_buy_agent_b['vwap'] = vwaps[i]
    dataset_buy_agent_s['vwap'] = vwaps[i]

    contador = 0
    for index, row in dataset_buy_agent_b.iterrows():
        if row['buy_agent'] == buy_agent and row['price'] < row['vwap']:
            contador += 1
        DatasetAnalizar.loc[(DatasetAnalizar['buy_agent'] == buy_agent), 'puntuacion'] = contador

    contador = 0
    for index, row in dataset_buy_agent_s.iterrows():
        if row['sell_agent'] == buy_agent and row['price'] > row['vwap']:
            contador += 1
        DatasetAnalizar.loc[(DatasetAnalizar['buy_agent'] == buy_agent), 'puntuacion'] = contador
    i += 1

DatasetAnalizar.head(50)
```
